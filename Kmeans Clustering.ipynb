```python
import matplotlib.pyplot as plt

# Sample data: Replace these with your actual SGPA grades
semesters = ['Sem 1', 'Sem 2', 'Sem 3', 'Sem 4', 'Sem 5', 'Sem 6']
sgpa_grades = [8.54, 7.8, 9.19, 9.52, 9.29, 9.65]
colors = ['red', 'blue', 'green', 'pink', 'purple', 'orange']

# Plotting the bar graph
fig, ax = plt.subplots(figsize=(7, 4))
bars = ax.bar(semesters, sgpa_grades, color=colors)

# Adding title and labels
ax.set_title('Semester-wise SGPA Grades')
ax.set_xlabel('Semesters')
ax.set_ylabel('SGPA')

# Adding text annotations for each bar using ax.annotate
for bar, sgpa in zip(bars, sgpa_grades):
    yval = bar.get_height()
    ax.annotate(f'{sgpa}', xy=(bar.get_x() + bar.get_width() / 2, yval),
                xytext=(0, -10),  # offset points
                textcoords='offset points',
                ha='center', va='bottom', color='black')

# Displaying the plot
plt.show()
```
